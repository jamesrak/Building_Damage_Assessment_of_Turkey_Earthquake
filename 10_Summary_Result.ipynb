```python
import pandas as pd
import json
import os
```

```python
base_dir = 'turkey_earthquake/'
submission_id = 'submission7'
result_dir = base_dir+'predict/evaluation-'+submission_id+'/'
sorted(os.listdir(result_dir))
```

```python
files = sorted(os.listdir(result_dir))
data_arr = []
for filename in files:
  file_path = result_dir+filename
  # Read the JSON file
  with open(file_path, "r") as json_file:
      data_dict = json.load(json_file)
      data_dict['city'] = filename.split('_')[1].split('.')[0]
      data_arr.append(data_dict)

df = pd.DataFrame(data_arr)
df
```

```python
df = df.round(3)
df = df[['city','score', 'damage_f1', 'localization_f1', 'damage_f1_no_damage',
       'damage_f1_minor_damage', 'damage_f1_major_damage',
       'damage_f1_destroyed']]
df
```

```python
df.to_csv(base_dir+'predict/evaluation-'+submission_id+'_turkey_earthquake.csv',index=False)
```

```python
# Get overall score
file_path = base_dir+'predict/overall_score_submission7-2.json'
# Read the JSON file
with open(file_path, "r") as json_file:
    data_dict = json.load(json_file)
    data_dict['city'] = 'Total'
df = pd.DataFrame([data_dict])
df = df.round(3)
df = df[['city','score', 'damage_f1', 'localization_f1', 'damage_f1_no_damage',
       'damage_f1_minor_damage', 'damage_f1_major_damage',
       'damage_f1_destroyed']]
df
df
```
